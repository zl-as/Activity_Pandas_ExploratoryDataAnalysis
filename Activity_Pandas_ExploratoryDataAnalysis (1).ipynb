```python
#What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) 
#and those who earn less than 50K per year?

ageMoreThan = []
ageLessThan = []

for i in range(0, len(df)):
    if (df.iloc[i][14] == ' >50K'):
        ageMoreThan.append(df.iloc[i][0])
    if (df.iloc[i][14] == ' <=50K'):
        ageLessThan.append(df.iloc[i][0])
        
print("Mean age of people earning more than 50K per year is", np.mean(ageMoreThan))
print("Standard Deviation of age for people earning more than 50K per year is", np.std(ageMoreThan))
print()
print("Mean age of people earning less than 50K per year is", np.mean(ageLessThan))
print("Standard Deviation of age for people earning less than 50K per year is", np.std(ageLessThan))

```

```python
#What is the maximum number of hours a person works per week (hours-per-week feature)? 
#How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?


maxhrs = df['hours-per-week'].max()
maxhrs

```
